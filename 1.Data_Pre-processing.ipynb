<h2 style="color:#08519c; border-bottom:3px solid #08519c; padding-bottom:4px; margin-top:15px;">
  3. Data Pre-processing
</h2>

#### <span style="color:#08519c">3.1  Handling missing values</span>

##### <font color='#0891b2'>Checking Missing Values:

```python
print(data.isnull().sum())
```

##### <font color='#0891b2'>For Numerical columns Handling missing values using Median: 

```python
data['experience_years'] = data['experience_years'].fillna(data['experience_years'].median())

data['skills_count']     = data['skills_count'].fillna(data['skills_count'].median())

data['certifications']   = data['certifications'].fillna(data['certifications'].median())
```

##### <font color='#0891b2'>For Categorical columns Handling missing values using Mode:

```python
data['education_level']  = data['education_level'].fillna(data['education_level'].mode()[0])

data['industry']         = data['industry'].fillna(data['industry'].mode()[0])

data['company_size']     = data['company_size'].fillna(data['company_size'].mode()[0])

data['location']         = data['location'].fillna(data['location'].mode()[0])

data['remote_work']      = data['remote_work'].fillna(data['remote_work'].mode()[0])
```

##### <font color='#0891b2'>Checking missing values are resolved:

```python
print("Missing values after handling:\n", data.isnull().sum())
```

#### <span style="color:#08519c">3.2  Removing Duplicates</span>

##### <font color='#0891b2'>checking for duplicate rows:

```python
duplicates_count= data.duplicated().sum()
print("Number of duplicate rows:",duplicates_count)
```

##### <font color='#0891b2'>Removing duplicate rows:

```python
data=data.drop_duplicates()
```

##### <font color='#0891b2'>Checking duplicate values resolved:

```python
duplicates_count= data.duplicated().sum()
print("Number of duplicate rows:",duplicates_count)
```

#### <span style="color:#08519c">3.3  Correcting data types</span>

##### <font color='#0891b2'>Converting numerical columns to integers:

```python
data['experience_years'] = data['experience_years'].astype('int64')
data['skills_count']     = data['skills_count'].astype('int64')
data['certifications']   = data['certifications'].astype('int64')
```

##### <font color='#0891b2'>Converting categorical columns to 'category':

```python
categorical_cols = ['job_title', 'education_level', 'industry',
                    'company_size', 'location', 'remote_work']

for col in categorical_cols:
    data[col] = data[col].astype('category')
```

##### <font color='#0891b2'>Verifying changes:

```python
print(data.dtypes)
```

#### <span style="color:#08519c">3.4  Creating derived columns</span>

##### <font color='#0891b2'>Experience group

```python
data['experience_group'] = pd.cut(
    data['experience_years'],
    bins=[0,5,10,15,20],
    labels=['Junior','Mid','Senior','Expert'])
```

##### <font color='#0891b2'>Skill level

```python
data['skill_level'] = pd.cut(
    data['skills_count'],
    bins=[0,5,10,15,20],
    labels=['Low','Medium','High','Very High'])
```

##### <font color='#0891b2'>Certification Status 

```python
data['has_certification'] = (data['certifications'] > 0).astype(int)
```

##### <font color='#0891b2'>Work Mode

```python
data['work_mode'] = data['remote_work'].map({
    'Yes': 'Remote',
    'No': 'On-Site',
    'Hybrid': 'Hybrid'
})
```

##### <font color='#0891b2'>Salary band

```python
data['salary_band'] = pd.cut(
    data['salary'],
    bins=[0,100000,150000,200000,350000],
    labels=['Low','Mid','High','Very High'])
```

##### <font color='#0891b2'>Verifying New columns added:

```python
print(data.columns)
```

#### <span style="color:#08519c">3.5  Filtering and aggregating data</span>

##### <font color='#0891b2'>Aggregating: count of employees by job title

```python
job_counts = data['job_title'].value_counts()

print(job_counts)
```

##### <font color='#0891b2'>Average salary by education level

```python
edu_salary = data.groupby('education_level', observed=False)['salary'].mean().round(2)
print(edu_salary)
```
